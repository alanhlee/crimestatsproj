# Dallas, TX

```python
import pandas as pd
dallas = pd.read_csv('Police_Incidents.csv', parse_dates=['Date1 of Occurrence'])
```

```python
dallas.columns = dallas.columns.str.lower().str.replace(" ", "_")
```

```python
dallas.columns
```

```python
dallas.head()
```

```python
dallas = dallas.rename(columns={'type_of_incident': 'crime_category', 'date1_of_occurrence': 'date'})
```

```python
dallas['city'] = 'Dallas'
```

```python
dallas = dallas.loc[:, ['city', 'date', 'crime_category', 'victim_race', 'victim_ethnicity', 'victim_gender', 'victim_age']]
```

```python
dallas = dallas[dallas.date >= '2020-10-08']
```

```python
dallas.to_csv('../Data_Clean/Dallas.csv')
```
