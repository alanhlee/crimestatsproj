# Nashville, TN

```python
import pandas as pd
nv = pd.read_csv('Metro_Nashville_Police_Department_Incidents.csv', parse_dates=['Incident Occurred'])
```

```python
nv.count()
```

```python
nv.head()
```

```python
nv.columns = nv.columns.str.lower().str.replace(" ", "_")
```

```python
nv = nv.rename(columns={'primary_key': 'city', 'incident_occurred': 'date', 'offense_description': 'crime_category'})
```

```python
nv['city'] = 'Nashville'
```

```python
nv.columns
```

```python
nv = nv.loc[:, ['city', 'date', 'crime_category', 'victim_gender', 'victim_race', 'victim_ethnicity']]
```

```python
nv = nv[nv.date >= '2020-10-08']
```

```python
nv.to_csv('../Data_Clean/Nashville.csv')
```
