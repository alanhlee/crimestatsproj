```python
#2020-10-02 last updated
import pandas as pd
cin = pd.read_csv('PDI__Police_Data_Initiative__Crime_Incidents.csv', parse_dates=['DATE_REPORTED'])
```

```python
cin.columns = cin.columns.str.lower().str.replace(" ", "_")
```

```python
cin.head()
```

```python
cin = cin.rename(columns={'instanceid': 'city', 'date_reported': 'date', 'offense': 'crime_category'})
```

```python
cin.columns
```

```python
cin = cin.loc[:, ['city', 'date', 'crime_category', 'victim_age', 'victim_race', 'victim_ethnicity', 'victim_gender', 'suspect_age', 'suspect_race', 'suspect_ethnicity', 'suspect_gender']]
```

```python
cin['city'] = 'Cincinnati'
```

```python
cin = cin[cin.date >= '2020-10-02']
```

```python
cin.to_csv('../Data_Clean/Cincinnati.csv')
```
