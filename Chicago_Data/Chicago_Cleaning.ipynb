```python
import pandas as pd
chicago = pd.read_csv("Crimes_-_2001_to_Present.csv", parse_dates=['Date'])
```

```python
chicago.columns = chicago.columns.str.lower().str.replace(" ", "_")
```

```python
chicago = chicago.rename(columns={'id': 'city', 'description': 'crime_category'})
```

```python
chicago['city'] = 'Chicago'
```

```python
chicago = chicago.loc[:, ['city', 'date', 'crime_category']]
```

```python
chicago = chicago[chicago.date >= '2020-09-25']
```

```python
chicago.to_csv('../Data_Clean/Chicago.csv')
```
