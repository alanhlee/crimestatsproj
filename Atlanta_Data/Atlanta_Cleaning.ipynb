# ATLANTA, GA

```python
import pandas as pd
import numpy as np
```

```python
atlanta_2021 = pd.read_csv("../Atlanta_Data/COBRA-2021.csv",
                 dtype={0: 'string'}, parse_dates={'occurred': [2, 3]})
#retrieved 7/7/21
```

```python
atlanta_2021.shape
atlanta_2021.dtypes
atlanta_2021.head()
atlanta_2021.count()
min(atlanta_2021.occurred)
max(atlanta_2021.occurred)
```

```python
atlanta_2020 = pd.read_csv("../Atlanta_Data/COBRA-2020(NEW RMS 9-30 12-31).csv",
                           dtype={0: "string"}, parse_dates={"occurred": [2, 3]})
#retrieved 7/9/21
```

```python
atlanta_2020.head()
atlanta_2020.shape
atlanta_2020.dtypes
atlanta_2020.count()
```

```python
#drop empty columns
atlanta_2020.dropna(how="all", axis=1, inplace=True)
```

```python
#drop crimes with no date and reset the index
atlanta_2020 = atlanta_2020[~atlanta_2020['occurred'].str.contains('nan')]
atlanta_2020.reset_index(drop=True, inplace=True)
```

```python
atlanta_2020[~atlanta_2020['occurred'].str.contains('2020')]
```

```python
atlanta_2020.loc[3599, 'occurred'] = '2020-10-30 17:56'
atlanta_2020.loc[4745, 'occurred'] = '2020-11-01 23:56'
```

```python
min(atlanta_2020.occurred)
max(atlanta_2020.occurred)
```

```python
#checking for differences in both dfs
atlanta_2020.columns.difference(atlanta_2021.columns)
```

```python
#combine dfs
#correct index first
start = atlanta_2020.shape[0]
stop = atlanta_2020.shape[0] + atlanta_2021.shape[0]
atlanta_2021.index = range(start, stop)
atlanta_list = [atlanta_2020, atlanta_2021]
atlanta = pd.concat(atlanta_list)
atlanta.shape
```

```python
atlanta.head()
```

```python
atlanta = atlanta.drop(columns=['offense_id', 'poss_date', 'poss_date', 'poss_time', 'beat', 'location', 'ibr_code', 'neighborhood', 'npu', 'long', 'lat', 'rpt'])
```

```python
atlanta = atlanta.rename(columns={'UC2_Literal': 'crime_category', 'rpt_date': 'date'})
```

```python
atlanta.head()
```

```python
atlanta.insert(loc=1, column='city', value='Atlanta')
```

```python
atlanta.head()
```

```python
atlanta = atlanta.drop(columns='occurred')
```

```python
atlanta
```

```python
atlanta['date'] = pd.to_datetime(atlanta['date'])
```
