# Denver, CO

```python
import pandas as pd
denver = pd.read_csv('crime.csv', parse_dates=['FIRST_OCCURRENCE_DATE'])
```

```python
denver.head()
```

```python
denver.count()
```

```python
denver.columns = denver.columns.str.lower().str.replace(" ", "_")
```

```python
denver = denver.rename(columns={'incident_id': 'city', 'offense_type_id': 'crime_category', 'first_occurrence_date': 'date'})
```

```python
denver['city'] = 'Denver'
```

```python
denver = denver.loc[:, ['city', 'date', 'crime_category']]
```

```python
denver = denver[denver.date >= '2020-10-08']
```

```python
denver.to_csv('../Data_Clean/Denver.csv')
```
