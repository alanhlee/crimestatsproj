```python
nyc_2020 = pd.read_csv(
"../NYC_Data/NYPD_Complaint_Data_Historic.csv", 
dtype={0: "string", 18: "string", 20: "string"}, 
#parse_dates={"complaint_from": [1, 2], "complaint_to": [3, 4]}
parse_dates=['RPT_DT', ["CMPLNT_FR_DT", "CMPLNT_FR_TM"], ["CMPLNT_TO_DT", "CMPLNT_TO_TM"]]
)

nyc_2020 = nyc_2020.rename(columns={"CMPLNT_FR_DT_CMPLNT_FR_TM": "complaint_from", "CMPLNT_TO_DT_CMPLNT_TO_TM":	"complaint_to"})

#retrieved 7/16/21
```

```python
nyc_2020.columns = nyc_2020.columns.str.lower()
```

```python
nyc_2020 = nyc_2020[nyc_2020['rpt_dt'] >= '2020-06-30']
```

```python
nyc_2020.dropna(how="all", axis=1, inplace=True)
```

```python
#check that both dfs have same columns
diff = nyc_2020.columns.difference(nyc_2021.columns)
diff
```

```python
#combine dfs
#correct index first
start = nyc_2020.shape[0]
stop = nyc_2020.shape[0] + nyc_2021.shape[0]
nyc_2021.index = range(start, stop)
nyc_list = [nyc_2020, nyc_2021]
nyc = pd.concat(nyc_list)
nyc.shape
```

```python
nyc.columns
```

```python
nyc['rpt_dt'] = pd.to_datetime(nyc['rpt_dt'])
```

```python
nyc.head()
```

```python
nyc = nyc.rename(columns={'complaint_to': 'city', 'ofns_desc': 'crime_category', 'rpt_dt': 'date'})
```

```python
nyc['city'] = 'New York City'
```

```python
nyc = nyc.loc[:, ['city', 'date', 'crime_category']]
```

```python
nyc.to_csv("../Data_Clean/NYC.csv")
```
