# Los Angeles, CA

```python
import pandas as pd
la = pd.read_csv('Crime_Data_from_2020_to_Present.csv', parse_dates=['Date Rptd'])
```

```python
la.head()
```

```python
la.columns = la.columns.str.lower().str.replace(" ", "_")
```

```python
la.count()
```

```python
la = la.rename(columns={'dr_no': 'city', 'date_rptd': 'date', 'crm_cd_desc': 'crime_category'})
```

```python
la['city'] = 'Los Angeles'
```

```python
la = la.loc[:, ['city', 'date', 'crime_category']]
```

```python
la = la[la.date >= '2020-10-19']
```

```python
la.to_csv('../Data_Clean/Los_Angeles.csv')
```
